```python
# ==========================================
# CÉLULA 0: Import dos Pacotes e Bibliotecas
# ==========================================

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
import scipy.stats as stats

# Configuração de estilo para os gráficos
sns.set_theme(style="whitegrid")
```

```python
# ==========================================
# CÉLULA 1: Geração de Dados e Carregamento
# ==========================================
lista_precos = [987.41, 584.76, 612.60, 668.29, 629.48, 933.20, 924.40, 740.76, 928.01, 523.74, 1105, 1244, 677.78]

# Criando o DataFrame
df = pd.DataFrame(lista_precos, columns=['Preco'])

# Visualizando as primeiras linhas
print(df.head())
print(f"Total de registros: {len(df)}")
print("\n" + "="*50 + "\n")
```

```python
# ==========================================
# CÉLULA 2: Cálculos Estatísticos Básicos
# ==========================================
print("--- CÉLULA 2: Estatísticas Descritivas ---")

media = df['Preco'].mean()
mediana = df['Preco'].median()
# A moda pode retornar múltiplos valores, pegamos o primeiro
moda = df['Preco'].mode()[0] 

print(f"Média:   {media:.2f}")
print(f"Mediana: {mediana:.2f}")
print(f"Moda:    {moda:.2f}")
print("\n" + "="*50 + "\n")
```

```python
# ==========================================
# CÉLULA 3: Boxplot (Identificação Visual) - IDENTIFICAÇÃO DE OUTLIERS
# ==========================================
print("--- CÉLULA 3: Gerando Boxplot ---")

plt.figure(figsize=(10, 6))
sns.boxplot(x=df['Preco'], color='skyblue')

# Adicionando linhas para Média, Mediana e Moda
plt.axvline(media, color='red', linestyle='--', label=f'Média ({media:.2f})')
plt.axvline(mediana, color='green', linestyle='-', label=f'Mediana ({mediana:.2f})')
# Nota: A moda nem sempre é visível perto da mediana em distribuições normais, mas está plotada
plt.axvline(moda, color='purple', linestyle=':', label=f'Moda ({moda:.2f})')

plt.title('Boxplot de Preços com Indicadores Estatísticos')
plt.xlabel('Preço ($)')
plt.legend()
plt.show() # Em um notebook, isso renderizaria o gráfico inline

print("Gráfico Boxplot gerado.")
print("\n" + "="*50 + "\n")
```

```python
# ==========================================
# CÉLULA 4: Gráfico de Distribuição Normal
# ==========================================
print("--- CÉLULA 4: Gráfico de Distribuição ---")

plt.figure(figsize=(10, 6))

# Histograma com KDE (Kernel Density Estimate - linha suave da distribuição)
sns.histplot(df['Preco'], kde=True, color='blue', bins=30, stat="density", label='Dados Reais')

# Plotando uma curva normal teórica para comparação
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = stats.norm.pdf(x, media, df['Preco'].std())
plt.plot(x, p, 'k', linewidth=2, label='Curva Normal Teórica')

plt.title('Distribuição dos Preços (Histograma + KDE vs Normal)')
plt.xlabel('Preço ($)')
plt.ylabel('Densidade')
plt.legend()
plt.show()

print("Gráfico de distribuição gerado.")
print("\n" + "="*50 + "\n")
```

```python
# ==========================================
# CÉLULA 5: Cálculo IQR e Eliminação de Outliers
# ==========================================
print("--- CÉLULA 5: Remoção de Outliers via IQR (Q3 - Q1) ---")

Q1 = df['Preco'].quantile(0.25)
Q3 = df['Preco'].quantile(0.75)
IQR = Q3 - Q1

limite_inferior = Q1 - 1.5 * IQR
limite_superior = Q3 + 1.5 * IQR

print(f"Q1 (25%): {Q1:.2f}")
print(f"Q3 (75%): {Q3:.2f}")
print(f"IQR (Q3-Q1): {IQR:.2f}")
print(f"Limites para não-outliers: [{limite_inferior:.2f}, {limite_superior:.2f}]")

# Filtrando o DataFrame
df_limpo = df[(df['Preco'] >= limite_inferior) & (df['Preco'] <= limite_superior)]

print(f"\nRegistros antes da limpeza: {len(df)}")
print(f"Registros após limpeza (sem outliers): {len(df_limpo)}")
print(f"Outliers removidos: {len(df) - len(df_limpo)}")

# Plot comparativo final (opcional, para visualização do resultado)
plt.figure(figsize=(10, 6))
sns.histplot(df_limpo['Preco'], kde=True, color='green')
plt.title('Distribuição de Preços Após Remoção de Outliers')
plt.xlabel('Preço ($)')
plt.show()
```

```python
# Convertendo para Pandas Series para facilitar manipulação estatística
series_precos = pd.Series(lista_precos)

print(f"Dados carregados. Total de observações: {len(series_precos)}")
```

```python
## Análises Após a Remoção de Outliers
# CÉLULA 2: Análise Inicial (Gráficos "Antes")
# ==============================================================================
print("\n--- Gerando Visualizações Iniciais ---")

# 1. Gráfico de Distribuição Normal (Histograma + Curva)
plt.figure(figsize=(10, 6))
# Histograma dos dados
sns.histplot(series_precos, kde=True, stat="density", color="skyblue", label="Dados Reais")

# Sobreposição da Curva Normal Teórica
media_inicial = series_precos.mean()
std_inicial = series_precos.std()
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = stats.norm.pdf(x, media_inicial, std_inicial)
plt.plot(x, p, 'k', linewidth=2, label='Curva Normal Teórica')

plt.title('Distribuição Inicial dos Preços')
plt.xlabel('Preço')
plt.legend()
plt.savefig('1_distribuicao_inicial.png') # Salvando em disco
plt.show()
print("-> Gráfico '1_distribuicao_inicial.png' salvo.")


# 2. Boxplot Vertical (Para identificar Outliers)
plt.figure(figsize=(6, 8))
sns.boxplot(y=series_precos, color="salmon")
plt.title('Boxplot Inicial (Identificação de Outliers)')
plt.ylabel('Preço')
plt.savefig('2_boxplot_inicial.png') # Salvando em disco
plt.show()
print("-> Gráfico '2_boxplot_inicial.png' salvo.")
```

```python
# CÉLULA 3: Cálculo de Quantis e Verificação de Outliers
# ==============================================================================
print("\n--- Verificando Estatísticas de Outliers (IQR) ---")

Q1 = series_precos.quantile(0.25)
Q3 = series_precos.quantile(0.75)
IQR = Q3 - Q1

limite_inferior = Q1 - (1.5 * IQR)
limite_superior = Q3 + (1.5 * IQR)

# Identificando quais são os outliers
outliers = series_precos[(series_precos < limite_inferior) | (series_precos > limite_superior)]
tem_outliers = len(outliers) > 0

print(f"Q1 (25%): {Q1:.2f}")
print(f"Q3 (75%): {Q3:.2f}")
print(f"IQR: {IQR:.2f}")
print(f"Faixa aceitável: {limite_inferior:.2f} até {limite_superior:.2f}")
print(f"\nQuantidade de Outliers detectados: {len(outliers)}")

if tem_outliers:
    print(f"Valores outliers: {outliers.values}")
else:
    print("Nenhum outlier detectado estatisticamente.")
```

```python
# CÉLULA 4: Remoção Condicional e Reavaliação (Pós-Limpeza)
# ==============================================================================
# Lógica condicional: Só executa a limpeza e novos gráficos se houver outliers
if tem_outliers:
    print("\n--- Removendo Outliers e Gerando Novos Gráficos ---")
    
    # Filtrando os dados
    series_limpa = series_precos[
        (series_precos >= limite_inferior) & 
        (series_precos <= limite_superior)
    ]
    
    # 3. Novo Gráfico de Distribuição Normal
    plt.figure(figsize=(10, 6))
    sns.histplot(series_limpa, kde=True, stat="density", color="green", label="Dados Limpos")
    
    # Nova Curva Normal baseada nos dados limpos
    mu_novo, std_novo = series_limpa.mean(), series_limpa.std()
    xmin, xmax = plt.xlim()
    x = np.linspace(xmin, xmax, 100)
    p = stats.norm.pdf(x, mu_novo, std_novo)
    plt.plot(x, p, 'k', linewidth=2, label='Nova Curva Normal')
    
    plt.title('Distribuição de Preços (Sem Outliers)')
    plt.xlabel('Preço')
    plt.legend()
    plt.savefig('3_distribuicao_limpa.png')
    plt.show()
    
    # 4. Novo Boxplot Vertical
    plt.figure(figsize=(6, 8))
    sns.boxplot(y=series_limpa, color="lightgreen")
    plt.title('Boxplot (Pós-Limpeza)')
    plt.ylabel('Preço')
    plt.savefig('4_boxplot_limpo.png')
    plt.show()
    
    print("-> Gráfico '3_distribuicao_limpa.png' salvo.")
    print("-> Gráfico '4_boxplot_limpo.png' salvo.")
    print(f"\nDados originais: {len(series_precos)} -> Dados finais: {len(series_limpa)}")

else:
    print("\nComo não foram detectados outliers, os gráficos originais representam a visão final.")
```

```python
# CÉLULA 5: Estatísticas Finais e Gráfico Detalhado
# ==============================================================================
print("\n--- Calculando Medidas de Posição Final ---")

# Cálculos nos dados finais (sejam eles limpos ou originais)
media_final = series_limpa.mean()
mediana_final = series_limpa.median()
moda_final = series_limpa.mode()[0] # Pega o primeiro valor caso seja multimodal

print(f"Média Final:   {media_final:.2f}")
print(f"Mediana Final: {mediana_final:.2f}")
print(f"Moda Final:    {moda_final:.2f}")

# Gráfico de Distribuição com Indicadores
plt.figure(figsize=(12, 7))
sns.histplot(series_limpa, kde=True, color="purple", stat="density", label='Distribuição Final')

# Linhas Verticais para Média, Mediana e Moda
plt.axvline(media_final, color='red', linestyle='--', linewidth=2, label=f'Média: {media_final:.2f}')
plt.axvline(mediana_final, color='orange', linestyle='-', linewidth=2, label=f'Mediana: {mediana_final:.2f}')
plt.axvline(moda_final, color='cyan', linestyle=':', linewidth=2, label=f'Moda: {moda_final:.2f}')

plt.title('Distribuição Final de Preços com Medidas de Posição')
plt.xlabel('Preço')
plt.ylabel('Densidade')
plt.legend()

nome_arquivo_stats = '5_distribuicao_estatisticas.png'
plt.savefig(nome_arquivo_stats)
plt.show()

print(f"-> Gráfico '{nome_arquivo_stats}' salvo com sucesso.")
```
